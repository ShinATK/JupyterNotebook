```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
accus_b0 = {}
accus_b3 = {}
for i in range(10):
    accus_b0[i] = np.loadtxt(f"accus{i}_ideal_ryx_b=0.txt")
    accus_b3[i] = np.loadtxt(f"accus{i}_ideal_ryx_b=3.txt")
    plt.plot(accus_b0[i], label=f"Ideal b=0 regnum={i}")
    plt.plot(accus_b3[i], label=f"Ideal b=3 regnum={i}")
    plt.xlim(0, 100)
    plt.ylim(0, 100)
    plt.legend()
    plt.show()
```

```python
accus_ideal = {}
accus_exp = {}
for i in range(10):
    accus_ideal[i] = np.loadtxt(f"accus{i}_ideal.txt")
    accus_exp[i] = np.loadtxt(f"accus{i}_exp.txt")
    plt.plot(accus_ideal[i], label=f"Ideal regnum={i}")
    plt.plot(accus_exp[i], label=f"Exp regnum={i}")
    plt.savefig(f"Ideal&Exp regnum={i}.png", dpi=720)
    plt.xlim(0, 100)
    plt.ylim(0, 100)
    plt.legend()
    plt.show()
```

```python
def GP(P, A=6, B=1, Pmin=0, Gmin=0):
    G = B * (1 - np.exp((Pmin-P) / A)) + Gmin
    return G
def GD(P, A=6, B=1, Pmax=50, Gmax=1):
    G = -B * (1 - np.exp((P-Pmax) / A)) + Gmax
    return G
P = np.arange(0, 51, 1)
basis = np.linspace(0, 1, 51)
plt.plot(P, basis, ls='--', c='black')
plt.plot(P, GP(P), label='GP')
plt.plot(P, GD(P), label='GD')
plt.legend()
plt.show()
```
